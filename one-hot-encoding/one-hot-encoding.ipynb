```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("homeprices.csv")
df
```

```python
y = pd.get_dummies(df.town, prefix='town')
df.drop(columns=['town'],inplace=True)
df = pd.concat([df,y],axis=1)
y_true = df.price
df.drop(columns=['price'],inplace=True)
```

```python
df
```

```python
model = linear_model.LinearRegression()
model.fit(df,y_true)
```

```python
model.predict(df)
```

```python
y_true
```

```python
model.score(df,y_true)
```

```python
df.drop(columns=['town_robinsville'],inplace=True);
model1 = linear_model.LinearRegression().fit(df,y_true)
```

```python
model1.score(df,y_true)
```
