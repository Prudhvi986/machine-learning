```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
```

```python
def gradientDescent(x,y,alpha):
    m_curr = b_curr = 0
    iterations=1000
    n = len(x)
    
    for i in range(iterations):
        y_predicted = m_curr*x + b_curr
        
        # plotting the current version
        plt.plot(x,y_predicted)
        cost  = np.mean((y-y_predicted) * (y-y_predicted))
        print("cost ",cost)
        delta_m = -(2/n)*sum(x*(y-y_predicted))
        delta_b = -(2/n)*sum((y-y_predicted))
        m_curr -= alpha * delta_m
        b_curr -= alpha * delta_b
```

```python
gradientDescent(x,y,0.1)
```

```python
gradientDescent(x,y,0.01)
```

```python
gradientDescent(x,y,0.001)
```

```python
gradientDescent(x,y,1)
```

Here we can clearly see that how gradient descent is working And how learning rate is effecting the model.
So we should choose a suitable learning rate.
if learning rate is high then it is very unstable, facing converging problem
if it is very low, then it takes lot of iterations to gain accuracy.
So learning rate should be selected optimal, by seeing the result of cost function.
